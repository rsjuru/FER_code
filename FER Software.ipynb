# FER Demo program using Py-Feat Toolbox

This code analyses facial expressions from two videoclips using Py-Feat Toolbox.
Funny animal videos were used as triggeres for positive facial expressions (happy).
In first video the facial expressions presented were spontanious reactions for funny video.
The second video tries to imitate the lack of emotional reactions in face caused by apathy.

Let's start by importing all needed libraries:

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
from feat import Detector
```

The detector is created by using Py-Feat Toolbox's pretrained detectors:

```python
detector = Detector(face_model="mtcnn", landmark_model="pfld", au_model="xgb", emotion_model="resmasknet", facepose_model="img2pose")
```

After that both video paths are determined:

```python
happy_video = os.path.join("video_happy.mp4")
apathy_video = os.path.join("video_apathy.mp4")
```

Prediction of the happy video is produced first:

```python
video_pred1 = detector.detect_video(happy_video,skip_frames=48,antialias=True)
video_pred1.head()
```

After the first prediction is produced few of the video frames are plotted and below that the emotions of the whole video are plotted:

```python
video_pred1.loc[[48]].plot_detections(faceboxes=True, add_titles=False)
```

```python
emotion_table = video_pred1.emotions
emotions1 = video_pred1.emotions.plot()
```

```python
video_pred2 = detector.detect_video(apathy_video,skip_frames=48,antialias=True)
video_pred2.head()
```

```python
video_pred2.loc[[48]].plot_detections(faceboxes=True)
```

```python
emotions2=video_pred2.emotions.plot()
```

```python
AUs1 = video_pred1.AU06.plot()
AUs2 = video_pred2.AU06.plot()
AU1 = video_pred1.AU12.plot()
AU2 = video_pred2.AU12.plot()
```

```python
AUs1 = video_pred1.aus.plot(title="AU predictions from spontaneous video")
AUs2 = video_pred2.aus.plot(title = "AU predictions from apathy video")
```
